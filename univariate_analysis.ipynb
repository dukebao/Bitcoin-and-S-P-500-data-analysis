Assuming we are using this past 4 year of data as our base line 
we will create the VaR ( Value at Risk ) statistic 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from math import *
```

```python
snpdata = pd.read_csv('S&P 500 Historical Data 2018 to 2022.csv')
snpdf = snpdata.set_index(pd.DatetimeIndex(snpdata['Date']))

# convert the data to float
snpdf['Price'] = snpdf['Price'].str.replace(',', '').astype(float)



# calcualte the confidence interval for S&P 500
SP_CI_mean = snpdf['Price'].mean()
SP_CI_std = snpdf['Price'].std()
SP_CI_n = snpdf['Price'].count()

SP_t_value = stats.t.ppf(1-0.05, SP_CI_n-1)
SP_CI_lower = SP_CI_mean - SP_t_value * SP_CI_std / sqrt(SP_CI_n)
SP_CI_upper = SP_CI_mean + SP_t_value * SP_CI_std / sqrt(SP_CI_n)
print("Note: 95% confidence interval")
print('S&P mean : ', SP_CI_mean)
print('S&P 500 Confidence Interval: ', SP_CI_lower, SP_CI_upper)
# normalize the data
print('S&P 500 normalized data')

bitcoindata = pd.read_csv('Bitcoin Historical Data - Investing.com 2018 to 2022.csv')
bitcoindf = bitcoindata.set_index(pd.DatetimeIndex(bitcoindata['Date']))

# convert the data to float
bitcoindf['Price'] = bitcoindf['Price'].str.replace(',', '').astype(float)

# calcualte the confidence interval for Bitcoin
BTC_CI_mean = bitcoindf['Price'].mean()
BTC_CI_std = bitcoindf['Price'].std()
BTC_CI_n = bitcoindf['Price'].count()

BTC_t_value = stats.t.ppf(1-0.05, BTC_CI_n-1)
BTC_CI_lower = BTC_CI_mean - BTC_t_value * BTC_CI_std / sqrt(BTC_CI_n)
BTC_CI_upper = BTC_CI_mean + BTC_t_value * BTC_CI_std / sqrt(BTC_CI_n)
print('bitcoin mean : ', BTC_CI_mean)
print('Bitcoin Confidence Interval: ', BTC_CI_lower, BTC_CI_upper)




```

What does this mean ? This means that based on the data for the past 4 years 

We are 95% confidence that the value of a unit of S&P500 will fall in between 3255 to 3321 

and we are 95% confidence that the value of a unit of Bitcoin will fall in between 17612 to 19141 





Howver , This does not look quiet right ...

The core issue is that Stock price is not normal distribution 

How do we fix it ? 

1. Find a distribution that matches the shape of your data and use that distribution to calculate the confidence interval.
2. Perform a transformation on your data to make it fit a normal distribution, and then find the confidence interval for the transformed data.

For this project , I will be transforming the data to make it fit a normal distribution 


```python
def map_x_with_data(x, data):
    print('x: ', x)
    print(data)
    return data.iloc[x]
```

```python
snpdata = pd.read_csv('S&P 500 Historical Data 2018 to 2022.csv')
snpdf = snpdata.set_index(pd.DatetimeIndex(snpdata['Date']))
x = np.array([i for i in reversed(range(len(snpdf['Price'])))])
snpdf['Price'] = snpdf['Price'].str.replace(',','').astype(float)
snpdf['Open'] = snpdf['Open'].str.replace(',','').astype(float)
snpdf['High'] = snpdf['High'].str.replace(',','').astype(float)
snpdf['Low'] = snpdf['Low'].str.replace(',','').astype(float)
plt.figure(figsize=(12, 6))
plt.plot(x, snpdf['Price'], color='blue', label='S&P 500')
plt.title('S&P 500 Price')
plt.xlabel('Date')
plt.ylabel('Price')


# Draw a straight line of best fit for S&P 500
# plt.figure(figsize=(12, 6))
x = np.array([i for i in reversed(range(len(snpdf['Price'])))])
# # convert the data to float
y = np.array(snpdf['Price'])
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='green', label='S&P 500')


plt.legend()
plt.show()
```

The green line represents the line of best fit 

and the deviation of the actual price from that line should now follow normal distribution

```python

# calculate the differce between the line and the data 
# for S&P 500

# print(snpdf['Price'])

# let y be the data
# let x be the index

calculated_y = m*x + b

# find out the actual value vs the calculated value
# for S&P 500
difference = y - calculated_y

# convernt array to numpy array 
difference = np.array(difference.tolist())

# caluclate the confidence interval for the difference
# for S&P 500
diff_CI_mean = difference.mean()
diff_CI_std = difference.std()
diff_CI_n = len(difference)
# diff_CI_z = 1.96
diff_t_value = stats.t.ppf(1-0.05, diff_CI_n-1)
diff_CI_lower = diff_CI_mean - diff_t_value * diff_CI_std / sqrt(diff_CI_n)
diff_CI_upper = diff_CI_mean + diff_t_value * diff_CI_std / sqrt(diff_CI_n)
print('S&P 500 Difference Confidence Interval: ', diff_CI_lower, diff_CI_upper)



```
